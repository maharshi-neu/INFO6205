```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display
%matplotlib inline
```

<h1>Insertion Sort Simulation</h1>
<h4> How was benchmarking done </h4>
<p>"x" is the size of the array selected (10 iterations) <br>
    Below cell contains raw run-time. Plotted on a loglog graph.</p>

```python
x = [128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536]

# raw values

y_random = [.17, .14, .50, 1.49, 5.99, 24.37, 99.28, 419.92, 1882.17, 8735.86]
plt.loglog(y_random, color='darkblue', label="Random")

y_reverse_range = [.17, .20, .78, 3.01, 11.38, 44.91, 178.07, 710.39, 2875.39, 11812.43]
plt.loglog(y_reverse_range, color='darkgrey', label="Reverse Range")

y_range = [.01, .01, .02, .09, .07, .06, .08, .24, .32, .63]
plt.loglog(y_range, color='orange', label="Range")

y_part_sorted = [.30, .15, .46, 1.51, 6.04, 25.62, 104.63, 454.92, 2041.21, 10227.92]
plt.loglog(y_part_sorted, color='yellow', label="Part Sorted")

plt.legend()

raw_df = pd.DataFrame()
raw_df['Size'] = x
raw_df['Range'] = y_range
raw_df['Part Sorted'] = y_part_sorted
raw_df['Random'] = y_random
raw_df['Reverse'] = y_reverse_range
display(raw_df)
display(raw_df.pct_change())
```

<h3>Best Case: Range <br>
Worse Case: Reverse Range</h3>

<p>Below cells contains normalized nlogn run-time.</p>

```python
x = [128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536]

y_reverse_range_nlogn = [39.25, 19.94, 33.52, 57.23, 96.90, 173.06, 313.44, 575.45, 1078.73, 2063.59]
y_range_nlogn = [3.21, 1.00, 1.03, 1.67, .59, .21, .14, .19, .12, .11]
y_random_nlogn = [39.51, 13.51, 21.53, 28.39, 50.99, 93.91, 174.75, 340.16, 706.11, 1526.12]
y_part_sorted_nlogn = [70.59, 14.65, 19.69, 28.61, 51.40, 98.75, 184.17, 368.51, 765.78, 1786.78]

plt.plot(x, y_reverse_range_nlogn, label="Reverse Range", color='darkgrey')
plt.plot(x, y_range_nlogn, label="Range", color='orange')
plt.plot(x, y_random_nlogn, label="Random", color='darkblue')
plt.plot(x, y_part_sorted_nlogn, label="Part Sorted", color='yellow')


plt.xlabel("SIZE")
plt.ylabel("TIME")
plt.legend()

nlogn_df = pd.DataFrame()
nlogn_df['Size'] = x
nlogn_df['Range'] = y_range_nlogn
nlogn_df['Part Sorted'] = y_part_sorted_nlogn
nlogn_df['Random'] = y_random_nlogn
nlogn_df['Reverse'] = y_reverse_range_nlogn
display(nlogn_df)
display(nlogn_df.pct_change())
```

<h2>Complexity of Insertion Sort / Order of Growth</h2>
<p>As seen above in the loglog plot insertion sort runs on average O(n) time in its best case and runs O(n^2) in its worst and average cases.</p>
